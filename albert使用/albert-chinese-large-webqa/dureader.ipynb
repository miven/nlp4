```python
def find_all(a, b):
    locs = []
    p = 0
    
    while True:
        q = a[p:].find(b)
        if q == -1:
            break
        else:
            locs.append(p + q)
            p = p + q + len(b)
    
    return locs
```

```python
def get_df(fn):
    rows = []
    
    with open(fn, 'r') as f:
        for line in tqdm(f):
            d = json.loads(line)
            question_id = 'dureader_' + str(d['question_id'])
            question = d['question']
            
            if len(d['answer_docs']) == 0 or len(d['fake_answers']) == 0 or d['answer_docs'][0] >= len(d['documents']):
                print('no answer doc!')
                continue

            doc = d['documents'][d['answer_docs'][0]]
            context = doc['paragraphs'][doc['most_related_para']]
            answer = d['fake_answers'][0]
            
            rows.append({
                'context': context,
                'question': question,
                'answer': answer
            })

    return pd.DataFrame(rows)
```

```python
df_train = pd.concat([
    get_df('data/dureader/preprocessed/trainset/zhidao.train.json'),
    get_df('data/dureader/preprocessed/trainset/search.train.json')],
    axis=0
)
df_train
```

```python
def get_data(fn):
    data = {
        "version": "v2.0",
        "data": [

        ]
    }
    
    with open(fn, 'r') as f:
        for line in tqdm(f):
            d = json.loads(line)
            question_id = 'dureader_' + str(d['question_id'])
            question = d['question']
            
            if len(d['answer_docs']) == 0 or len(d['fake_answers']) == 0 or d['answer_docs'][0] >= len(d['documents']):
                print('no answer doc!')
                continue

            doc = d['documents'][d['answer_docs'][0]]
            context = re.sub(r'\s+', ' ', doc['paragraphs'][doc['most_related_para']][:1000])
            answer = re.sub(r'\s+', ' ', d['fake_answers'][0])
            
            finds = find_all(context, answer)
            
            if len(finds) == 0:
                context = re.sub(r'\s+', '', context)
                answer = re.sub(r'\s+', '', answer)
                finds = find_all(context, answer)
                
            if len(finds) == 0:
                print('no finds', answer)
                print(context)
                continue
                
            qas = [{
                        'question': question,
                        'id': question_id,
                        'answers': [
                            {
                                'text': answer,
                                'answer_start': f
                            } for f in finds
                        ],
                        'is_impossible': False,
                    }]
                
            
            cqa = {
                'title': question_id,
                'paragraphs': [{
                    'context': context,
                    'qas': qas
                }]
            }
            data['data'].append(cqa)

    return data
```

```python
train_data = get_data('data/dureader/preprocessed/trainset/zhidao.train.json')
train_data
```
